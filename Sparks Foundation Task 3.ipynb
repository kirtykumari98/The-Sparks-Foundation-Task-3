For the given 'Iris' datset, create the Decision Tree classifier and visualize it graphically.

# Importing the libraries

```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

# Importing the dataset

```python
#importing the dataset
iris = datasets.load_iris()
X= iris.data
y= iris.target
```

```python
#no of attributes present in our dataset
iris.feature_names
```

```python
#catergories present in our dataset
iris.target_names
```

```python
X.shape,y.shape
```

# Fitting the dataset in Decision Tree Classifier model

```python
#fitting the dataset in decision tree classifier
classifier = DecisionTreeClassifier()
model = classifier.fit(X, y)
```

# Visualising the Decision Tree 

```python
fig= plt.figure(figsize=(20,20))
Fig= tree.plot_tree(classifier,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```
